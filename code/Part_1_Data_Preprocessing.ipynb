```python
# Define the JSON file path
json_file = 'yelp_academic_dataset_business.json'

# Initialize an empty list to hold chunks of data
data_chunks = []

# Define the chunk size (adjust as needed)
chunk_size = 100000

# Iterate over chunks of the JSON file
for chunk in pd.read_json(os.path.join(path, json_file), lines = True, chunksize=chunk_size):
    # Process each chunk of data (e.g., append to a list)
    data_chunks.append(chunk)

# Concatenate all chunks into a single DataFrame
business_df = pd.concat(data_chunks, ignore_index=True)
business_df # name, state, stars, review_count, attributes, categories
```

```python
business_df[business_df['is_open'] == 1]['categories']
```

```python
business_df.isnull().sum()
```

```python
open_business_with_categories = business_df[(business_df['is_open'] == 1) & (business_df['categories'].notnull())]
open_business_with_categories
```

```python
open_business_with_categories.describe()
```

```python
restaurant_df = open_business_with_categories[open_business_with_categories['categories'].str.contains('restaurants', case=False)]
restaurant_df
```

```python
restaurant_df[restaurant_df['state'] == 'CA']['city'].value_counts()
```

```python
restaurant_df.describe()
```

```python
restaurant_filtered = restaurant_df[['business_id', 'name', 'state', 'stars', 'review_count', 'attributes', 'categories']]
restaurant_filtered
```

```python
# Rename the 'stars' column in restaurant_df to 'stars_business'
restaurant_filtered = restaurant_filtered.rename(columns={'stars': 'restaurant_stars'})

# Rename the 'stars' column in yelp_df to 'stars_user'
yelp_df = yelp_df.rename(columns={'stars': 'user_stars'})

# Merge the DataFrames on 'business_id'
merged_df = pd.merge(restaurant_filtered, yelp_df, on='business_id', how='inner')
merged_df
```
